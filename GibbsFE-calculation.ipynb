```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.constants import Boltzmann
from scipy.constants import Avogadro
import math
import pandas as pd
def import_columns_from_xvg(xvg_file, columns):
    # Load the data from the XVG file
    data = np.loadtxt(xvg_file, comments=['#', '@'], usecols=columns)
    
    # Transpose the data to get each column as a separate array
    arrays = np.transpose(data)
    
    return arrays

T= 393 #K 

def gibbs_fe(T, density):
    
    gibbs= - Boltzmann * T * math.log(density)

    return gibbs
```

```python
T_low=310
def gibbs_low(T_low, density):
    
    gibbs= - Boltzmann * T_low * math.log(density)

    return gibbs

T_high=440
def gibbs_high(T_high, density):
    
    gibbs= - Boltzmann * T_low * math.log(density)

    return gibbs
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_first_last_columns(csv_file):
    # Read the CSV file
    df = pd.read_csv(csv_file)
    
    # Ensure there are at least two columns
    if df.shape[1] < 2:
        print("Error: The CSV file must have at least two columns.")
        return
    
    # Extract the first and last columns
    first_column = df.iloc[:, 0]
    last_column = df.iloc[:, -1]
    
    # Plot the first column against the last column
    plt.figure(figsize=(8, 6))
    plt.plot(first_column, last_column, 'o-', label=f'{df.columns[0]} vs {df.columns[-1]}')
    plt.xlabel(df.columns[0])
    plt.ylabel(df.columns[-1])
    plt.title('Plot of First Column vs Last Column')
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
plot_first_last_columns('/media/favour/LaCie/collab/cmld/plot1d/p_delG_Z.csv')
plot_first_last_columns('/media/favour/LaCie/collab/zota/plot1d/p_delG_Z.csv')
```
